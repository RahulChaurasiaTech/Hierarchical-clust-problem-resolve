```python
import pandas as pd
import numpy as np
import seaborn as sns
import matplotlib.pyplot as plt

tra= pd.read_csv("C:\Users\Lakshya\Downloads\google_review_ratings.csv")
tra.head()
```

```python
tra.columns
```

```python
tra.drop(columns=["User"], inplace= True)
```

tra.head()

```python
tra.head()
```

```python
from sklearn.preprocessing import normalize
tra_scaled = normalize(tra)
tra_scaled = pd.DataFrame(tra_scaled, columns=tra.columns)
tra_scaled.head()
```

```python
import scipy.cluster.hierarchy as shc
```

```python
import scipy.cluster.hierarchy as shc
plt.figure(figsize=(10, 7))  
plt.title("Dendrograms") 
dend = shc.dendrogram(shc.linkage(tra_scaled, method='ward'))
```

```python
from sklearn.cluster import AgglomerativeClustering

cluster = AgglomerativeClustering(n_clusters=5, affinity='euclidean', linkage='ward')  
```

```python
cl=cluster.fit_predict(tra_scaled)
```

```python
tra["cluster"]= cl
```

```python
tra.head()
```

```python
fg=tra.groupby(["cluster"]).mean().T
fg
```

```python
fg.to_csv("C:\Users\Lakshya\Documents\New folder\hier clus report.csv")
```

```python
profil0= pd.DataFrame(tra.groupby(["cluster"]).mean().T)
profil0.columns= ["cluster0", "cluster1", "cluster2", "cluster3", "cluster4"]
profil0.reset_index(inplace= True)
profil0
```

```python
profil1= pd.DataFrame(tra.describe().T["mean"])
profil1.columns= ["overall"]
profil1.reset_index(inplace= True)

final_profile= pd.merge(profil0, profil1, on= "index", how= "inner")
final_profile
```

```python
final_profile.to_csv("C:\Users\Lakshya\Documents\New folder\hier2.csv")
```
